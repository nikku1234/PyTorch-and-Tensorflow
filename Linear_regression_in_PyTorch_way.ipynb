```
import torch 
from torch.autograd import Variable
import numpy as np
```

```
x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0]]))
y_data = Variable(torch.Tensor([[2.0],[4.0],[6.0]]))
```

```
class Model(torch.nn.Module):
  def __init__(self):
    #Here in this constructor we instantiate two nn.Linear module
    super(Model,self).__init__()
    self.linear = torch.nn.Linear(1,1) # one in and one out

  def forward(self,x):
    # Accept a variable of input data snd return a variable of output data.
    y_pred = self.linear(x)
    return y_pred

#model
model = Model()
```

```
#Constructing a loss function and optimizer 

criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(),lr=0.01) #model.parameters will contain the learnable parameters of the two nn.Linear modules are members of this model
```

```
#Training loop

for epoch in range(500):
  #forward pass compute predicted y by passing x to the model
  y_pred = model(x_data)

  #compute and printing loss
  loss = criterion(y_pred,y_data)
  print(epoch, loss)

  #zero gradient, perform a backward pass, and update the weights
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```
#test
hour_var = Variable(torch.Tensor([[4.0]]))
print("Predict (after training)",4,model.forward(hour_var).data[0][0])
```
